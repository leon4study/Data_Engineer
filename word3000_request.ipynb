```python
import pandas as pd
import os, requests 
from bs4 import BeautifulSoup as bs

url = 'https://www.ef.com/wwen/english-resources/english-vocabulary/top-3000-words/'
selector = '#main-content > div > div > section > div > div > p'

page = requests.get(url)

if page.status_code == 200:
    soup = bs(page.text, "html.parser")
    
    elements = soup.select(selector=selector)
```

```python
text = [word for word in elements[0].stripped_strings]
len(text)
```

```python
text[7:14]
```

```python
import random

shuffled_text = random.sample(text, len(text))
shuffled_text
```

```python
chunk_size = 7

chunked_list = []

for i in range(0, len(shuffled_text), chunk_size):
    chunked_list.append(shuffled_text[i:i+chunk_size])
```

```python
chunked_list
```

```python
df = pd.DataFrame(chunked_list)
os.chdir("/Users/jun/Desktop/Eng/reference")
df.to_excel("top_3000_words.xlsx")
```
